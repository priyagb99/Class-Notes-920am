```python
left_df.merge(ChosenOne,
              how='inner',left_on='firm',right_on='tic') # 3 long

# 1:m --> one obs becomes duplicated with info in right for matching keys
# left dataset grows!
```

```python
CurryForThree.merge(left_df,
              how='inner',right_on='firm',left_on='var1') # 3 long

# M:1 --> each of the many get matched to one!
# dataset does NOT grow
```

```python
ChosenOne.merge(CurryForThree,
              how='inner',left_on='tic',right_on='var1') # 9 long!

# m:m --> each row can multiply!
```

```python
left_df.merge(ChosenOne,
              how='left',left_on='firm',right_on='tic') # 

# 1:m, with outer --> keep all unmatched obs from each
```
